```python
new_catID = conn.execute('select cat_id from categories where cat_name = (?)' , 
                         ['new_category']).fetchall()[0][0]
new_books = [['book1' , 33 , 3 ,new_catID],['book2' , 44 , 4,new_catID],
             ['book3' , 55 , 5 ,new_catID],['book4' , 22 , 2 ,new_catID],
             ['book5' , 11 , 1,new_catID]]
qry = 'insert into books (book_name , book_price , book_rate , cat_id) values(?,?,?,?)'
for book in new_books :
    conn.execute(qry , book)
pd.read_sql('select * from books' , conn)
```

```python
14) Write a sql query to join both books & categories table into one new table containing book_name & category_name & book_rate and book_price.
```

```python
qry = '''SELECT categories.cat_name , books.book_name , books.book_price , books.book_rate
         from categories , books
         where categories.cat_ID = books.cat_ID;'''
data = pd.read_sql(qry , conn)
data
```

```python
15) Write a sql query to calculate how many books each rate has and have price between 20£ and 30£.
```

```python
qry = '''SELECT book_rate , count(book_id) as num_of_books from books
         where book_price between 20 and 30 
         GROUP by book_rate;'''
data = pd.read_sql(qry , conn)
data
```

```python
16) Write a sql query to calculate how many books each category has having count > 10.
```

```python
qry = '''SELECT  categories.cat_name category_name , count(books.book_id) num_of_books
         from books , categories
         WHERE categories.cat_ID = books.cat_ID
         GROUP by books.cat_ID
         HAVING count(books.book_id) > 10;'''
data = pd.read_sql(qry , conn)
data
```

```python
17) Write a sql query to get all books with category_name='Music' using subquery.
```
